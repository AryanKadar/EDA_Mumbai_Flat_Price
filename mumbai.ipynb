```python
import pandas as p
import seaborn as s
import matplotlib.pyplot as m
p.set_option("display.max_columns",None)
p.set_option("display.max_rows",None)
t=p.read_csv("Mumbai.csv")
```

```python
t.head()
```

```python
m.subplots(figsize=(10,8))
k=t.groupby(["Area","Location"])["Price"].median().plot(rot=45)
m.xlabel("Location",fontsize=15);
m.ylabel("Price",fontsize=15);
m.xticks(size=12);
```

```python
t.groupby("No. of Bedrooms")["Price"].median().plot(marker="v",figsize=(10,8))
m.ylabel("Price",fontsize=15);
m.xlabel("No. of Bedrooms",fontsize=14);
```

```python
ko=['New/Resale','Gymnasium','Lift Available','Car Parking','Maintenance Staff','24x7 Security',"Children's Play Area",'Clubhouse','Intercom','Landscaped Gardens','Indoor Games','Gas Connection','Jogging Track','Swimming Pool']
```

```python
#f,a=m.subplots(7,2)
for i,j in enumerate(ko):
    m.figure(i)
    t.groupby(j)["Price"].median().plot(kind="pie",autopct="%.f%%",startangle=290,radius=1)
    m.ylabel("")
    #m.xlabel(j,fontsize=12);
    m.title(j)
```

```python
t.boxplot(column="Price",vert = 0,figsize=(15,8));
```

```python
f,fg=m.subplots(figsize=(15,8))
s.distplot(t["Price"])
```

```python
k=t.iloc[:,2:].corr()
k=k[k>0.5]
f,a=m.subplots(figsize=(15,10))
s.heatmap(k,annot=True,cmap="Blues",linewidths = 2.5, linecolor = 'black');
```
